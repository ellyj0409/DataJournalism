```python
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import time

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
#비밀남녀 재생목록 KBSN <클립 모아보기>
driver.get("https://www.youtube.com/playlist?list=PLP9SGK5OdynaF6PQT2JPBUgAwVOTSxn11")
```

```python
contentsList = []

last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    
    time.sleep(3)
    
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    if new_height == last_height:
        break
        
    last_height = new_height
    
page = driver.page_source
soup = BeautifulSoup(page, 'lxml')

for i in range(132):
    title = soup.find_all('a', 'yt-simple-endpoint style-scope ytd-playlist-video-renderer')[i]['title']
    url = soup.find_all('a', 'yt-simple-endpoint style-scope ytd-playlist-video-renderer')[i]['href']

    contentsList.append(
    {'영상제목' : title,
     'url' : url, 
     '업로드날짜' : 0,
     '조회수' : 0,
     '댓글수' : 0})

    print(i, title, url)
```

```python
contentsList = pd.DataFrame(contentsList)
contentsList
```

```python
from selenium.webdriver.common.by import By
```

```python
# 날짜, 댓글 수, 조회 수 크롤링

for i in range(len(contentsList)):
    base_url = "https://www.youtube.com"
    url_id = contentsList['url'][i]
    url = base_url+url_id
    
    driver.get(url)
    
    time.sleep(5)
    
    # 댓글 개수 페이지 소스 로딩 위해 스크롤 살짝 내림 
    driver.execute_script("window.scrollTo(0, 300);")
    
    time.sleep(20)
    
    page = driver.page_source
    soup = BeautifulSoup(page, 'lxml')
    
    # 영상 게시 날짜
    date_text = soup.find_all('yt-formatted-string', 'style-scope ytd-video-primary-info-renderer')[1].string
    date_text = date_text.replace('실시간 스트리밍 시작일:', '').replace('.','').replace(' ','')
    date = datetime.strptime(date_text, '%Y%m%d').date()
    
    # 댓글 수
    try :
        comments_text = soup.select('#count > yt-formatted-string > span')[1].string
        comments = int(comments_text.replace(',',''))
    except:
        comments = "댓글사용중지"
    
    # 조회 수
    view_text = soup.find('span', 'view-count style-scope ytd-video-view-count-renderer').text
    view = int(view_text.replace('조회수 ', '').replace('회', '').replace(',',''))
    
    contentsList['업로드날짜'][i] = date
    contentsList['조회수'][i] = view
    contentsList['댓글수'][i] = comments
    
    print(i, contentsList['영상제목'][i], date, view, comments)
```

```python
# 댓글 크롤링 코드(댓글 많은 영상)

url = "https://www.youtube.com/watch?v=0nJbzXWkFKY&list=PLP9SGK5OdynaF6PQT2JPBUgAwVOTSxn11&index=81"

driver.get(url)

# 길이가 짧은 영상은 댓글 수집 전 다음 영상으로 넘어가므로 영상 멈춤
interval = 2
driver.find_element(By.XPATH, r'//*[@id="movie_player"]').click()
time.sleep(interval)
    
height = 0
while height <= 100000:
    height += 500
    driver.execute_script('window.scrollTo(0, ' +str(height)+ ')')
    time.sleep(2)
    
driver.execute_script("window.scrollTo(0, 500);")
time.sleep(2)
    
page = driver.page_source
soup = BeautifulSoup(page, 'lxml')

comments_list = []
comments = soup.select('yt-formatted-string#content-text')
for i in range(len(comments)):
    comments1 = comments[i].text
    comments1.replace('\n','').replace('\t','')
    comments_list.append(comments1)

print(comments_list)
```
