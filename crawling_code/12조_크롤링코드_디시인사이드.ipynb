```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
path = './chromedriver' 
num_list = []

# 기본 환승연애 갤러리 개념글 페이지 URL
BASE_URL = "https://gall.dcinside.com/mgallery/board/lists/?id=transferlove&list_num=100&sort_type=N&exception_mode=recommend&search_head=&page="

# 개념글 내 게시물 고유 번호 크롤링
for i in range(1, 38):
    search_url = BASE_URL+str(i)
    driver = webdriver.Chrome(path)
    driver.get(search_url) 
    time.sleep(3) 
    
    html = driver.page_source
    driver.quit()
    
    soup = BeautifulSoup(html, "html.parser")
    word = soup.find_all("td", class_="gall_num")
    for w in range(0, len(word)):
        if word[w].text.isdigit(): 
            num_list.append(word[w].text)
```

```python
# 크롤링 된 고유 번호
num_list
```

```python
import random
# 크롤링한 번호 중 1000개만 추출 (실제 데이터 크롤링에 지나치게 많은 시간이 소요되어 1천개로 축소)
rand_num_list = random.sample(num_list, 1000)
rand_num_list
```

```python
# 크롤링 속도 향상을 위한 설정
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
# headless 설정
chrome_options.headless = True

from selenium.webdriver.common.desired_capabilities import DesiredCapabilities

caps = DesiredCapabilities().CHROME
caps["pageLoadStrategy"] = "none"
```

```python
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
from selenium import webdriver
import time
from selenium.webdriver.common.by import By

# 크롤링한 게시물 고유번호 바탕으로 각 게시물 내 정보 크롤링
path = './chromedriver' 
POST_URL = 'https://gall.dcinside.com/board/view/?id=transferlove&no='
df_final2 = pd.DataFrame(columns=['date', 'title', 'content', 'views', 'likes'])

n = 0
# 한번에 1천개씩 돌리면 멈추는 경우가 잦아, 300~400개 가량씩 나눠서 크롤링
for num in tqdm(rand_num_list[601:]):
    try:
        search_url = POST_URL+num
        driver = webdriver.Chrome(path)
        driver.get(search_url)
        time.sleep(3)
        html = driver.page_source
        driver.quit()

        soup = BeautifulSoup(html, "html.parser")
        date = soup.find("span", class_="gall_date")
        title = soup.find("span", class_="title_subject")
        content = soup.find("div", class_="write_div")
        views = soup.find("span", class_="gall_count")
        likes = soup.find("span", class_="gall_reply_num")

        df_final2.loc[n, 'date'] = date.text
        df_final2.loc[n, 'title'] = title.text
        df_final2.loc[n, 'content'] = content.text
        df_final2.loc[n, 'views'] = views.text[3:]
        df_final2.loc[n, 'likes'] = likes.text[3:]
        n += 1
    except:
        pass
```
